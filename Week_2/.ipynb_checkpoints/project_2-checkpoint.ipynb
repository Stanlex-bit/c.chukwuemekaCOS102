```python
import math
import numpy as np


def quadratic():
    a = float(input("Input value for A: "))
    b = float(input("Input value for B: "))
    c = float(input("Input value for C: "))


    det = b**2 - 4*a*c


    if det > 0:
        x1 = (-b + math.sqrt(det))/(2*a)
        x2 = (-b - math.sqrt(det))/(2*a)
        print(f"The roots of the equation are {x1} and {x2}.")
    elif det == 0:
        x = -b / (2*a)
        print(f"The root of your equation is {x1}.")
    elif det < 0:
        print(" There are no real roots")


def cubic():
    a = float(input("Input value for A: "))
    b = float(input("Input value for B: "))
    c = float(input("Input value for C: "))
    d = float(input("Input value for D: "))


    coefficients = [a, b, c, d]
    roots = np.roots(coefficients)
    print("Roots of the cubic equation are: ", roots)


def quartic():
    a = float(input("Input value for A: "))
    b = float(input("Input value for B: "))
    c = float(input("Input value for C: "))
    d = float(input("Input value for D: "))
    e = float(input("Input value for E: "))


    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)
    print("Roots of the quartic equation are: ", roots)


print("Equation solver \n")
print("""Key codes  
Quadratic equation-1 
Cubic equation-2 
Quartic equation-3""")
first_i = int(input("What type of equation do you want to solve(using key codes)? "))


if first_i == 1:
    quadratic()
elif first_i == 2:
    cubic()
elif first_i == 3:
    quartic()
else:
    print("Wrong input")

```

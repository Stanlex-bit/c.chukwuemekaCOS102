```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

def button_click():
    # Show an information message box
    messagebox.showinfo("Info", "Welcome to COS 102 GUI App!\n")
    
    # Ask for user confirmation
    result = messagebox.askyesno("Confirmation", "Do you want to continue?")
    
    if result:
        # If user clicks Yes, proceed to verification
        check_employee()
    else:
        # If user clicks No, show goodbye message
        messagebox.showinfo("Goodbye", "Thank you for using the app!")

# Load the dataset
try:
    df = pd.read_csv('GIG-logistics.csv')
    # Strip whitespace from columns to avoid matching errors
    df.columns = df.columns.str.strip()
except FileNotFoundError:
    print("Error: GIG-logistics.csv not found!")

def check_employee():
    name_input = entry_name.get().strip()
    dept_input = entry_dept.get().strip()
    
    # Convert to lowercase for case-insensitive comparison
    name_lower = name_input.lower()
    dept_lower = dept_input.lower()
    
    # Create temporary lowercase columns for matching
    df['Name_lower'] = df['Name'].str.lower()
    df['Dept_lower'] = df['Department'].str.lower()
    
    # Search for the employee
    employee = df[(df['Name_lower'] == name_lower) & 
                  (df['Dept_lower'] == dept_lower)]
    
    if not employee.empty:
        # Get the actual name and department from the dataset
        actual_name = employee.iloc[0]['Name']
        actual_dept = employee.iloc[0]['Department']
        
        # Welcome the employee
        welcome_msg = f"Welcome, {actual_name}!\n\n"
        
        # Get other members of the same department
        colleagues = df[df['Dept_lower'] == dept_lower]['Name'].tolist()
        colleagues.remove(actual_name) # Remove the current user from the list
        
        colleagues_str = ", ".join(colleagues) if colleagues else "None"
        messagebox.showinfo("Success", f"{welcome_msg}Your colleagues in {actual_dept}: \n{colleagues_str}")
    else:
        messagebox.showerror("Error", "Employee does not exist in our records.")
    
    # Clean up temporary columns
    df.drop(['Name_lower', 'Dept_lower'], axis=1, inplace=True)

# GUI Setup
root = tk.Tk()
root.title("GIG Logistics Verification")
root.geometry("300x200")

tk.Label(root, text="Enter Name:").pack(pady=5)
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Enter Department:").pack(pady=5)
entry_dept = tk.Entry(root)
entry_dept.pack()

tk.Button(root, text="Verify Identity", command=button_click).pack(pady=20)

root.mainloop()
```

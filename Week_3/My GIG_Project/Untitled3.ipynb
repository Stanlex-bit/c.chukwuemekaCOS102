```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import os

# Load the dataset with error handling
try:
    # Check if file exists
    if not os.path.exists('GIG-logistics.csv'):
        messagebox.showerror("Error", "GIG-logistics.csv file not found!")
        exit()
    
    df = pd.read_csv('GIG-logistics.csv')
    # Strip whitespace from columns to avoid matching errors
    df.columns = df.columns.str.strip()
    
    # Strip whitespace from all string data
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.strip()
    
    print(f"Loaded {len(df)} employees successfully")  # Debug print
    
except Exception as e:
    print(f"Error loading file: {e}")
    messagebox.showerror("Error", f"Failed to load data: {e}")

def check_employee():
    # Get input and strip whitespace
    name_input = entry_name.get().strip()
    dept_input = entry_dept.get().strip()
    
    # Validate input
    if not name_input or not dept_input:
        messagebox.showwarning("Warning", "Please enter both name and department!")
        return
    
    # Convert inputs to lowercase for case-insensitive matching
    name_lower = name_input.lower()
    dept_lower = dept_input.lower()
    
    # Search for the employee (case-insensitive)
    # Create temporary columns with lowercase for comparison
    df['Name_lower'] = df['Name'].str.lower()
    df['Dept_lower'] = df['Department'].str.lower()
    
    employee = df[(df['Name_lower'] == name_lower) & 
                  (df['Dept_lower'] == dept_lower)]
    
    if not employee.empty:
        # Get the actual name as it appears in the dataset (for display)
        actual_name = employee.iloc[0]['Name']
        actual_dept = employee.iloc[0]['Department']
        
        # Welcome message
        welcome_msg = f"Welcome, {actual_name}!\n"
        welcome_msg += f"Department: {actual_dept}\n"
        welcome_msg += f"Role: {employee.iloc[0]['Role']}\n\n"
        
        # Get other members of the same department
        colleagues = df[df['Dept_lower'] == dept_lower]['Name'].tolist()
        
        # Remove the current user from the list
        current_name = employee.iloc[0]['Name']
        if current_name in colleagues:
            colleagues.remove(current_name)
        
        # Format the colleagues list
        if colleagues:
            colleagues_str = "\n".join([f"• {colleague}" for colleague in colleagues])
            colleagues_msg = f"Other employees in {actual_dept} department:\n{colleagues_str}"
        else:
            colleagues_msg = f"You are the only employee in the {actual_dept} department."
        
        # Show success message
        messagebox.showinfo("Verification Successful", f"{welcome_msg}{colleagues_msg}")
    else:
        # Politely indicate employee doesn't exist
        messagebox.showerror("Verification Failed", 
                            f"Sorry, we could not find '{name_input}' in the {dept_input} department.\n\n"
                            f"Please check your spelling or contact HR for assistance.")
    
    # Clean up temporary columns
    df.drop(['Name_lower', 'Dept_lower'], axis=1, inplace=True, errors='ignore')

def clear_fields():
    entry_name.delete(0, tk.END)
    entry_dept.delete(0, tk.END)

# GUI Setup with better styling
root = tk.Tk()
root.title("GIG Logistics Employee Verification System")
root.geometry("400x300")
root.resizable(False, False)

# Add a header label
header_label = tk.Label(root, text="GIG LOGISTICS", font=("Arial", 16, "bold"), fg="blue")
header_label.pack(pady=10)

subheader_label = tk.Label(root, text="Employee Verification Portal", font=("Arial", 10))
subheader_label.pack(pady=(0, 20))

# Create a frame for better organization
main_frame = tk.Frame(root)
main_frame.pack(pady=10)

# Name input
tk.Label(main_frame, text="Employee Name:", font=("Arial", 10)).grid(row=0, column=0, sticky="w", pady=5, padx=5)
entry_name = tk.Entry(main_frame, width=30, font=("Arial", 10))
entry_name.grid(row=0, column=1, pady=5, padx=5)
entry_name.focus()  # Set focus to name field

# Department input
tk.Label(main_frame, text="Department:", font=("Arial", 10)).grid(row=1, column=0, sticky="w", pady=5, padx=5)
entry_dept = tk.Entry(main_frame, width=30, font=("Arial", 10))
entry_dept.grid(row=1, column=1, pady=5, padx=5)

# Buttons frame
button_frame = tk.Frame(root)
button_frame.pack(pady=20)

# Verify button
verify_btn = tk.Button(button_frame, text="Verify Identity", command=check_employee, 
                       bg="green", fg="white", font=("Arial", 10), padx=20, pady=5)
verify_btn.grid(row=0, column=0, padx=10)

# Clear button
clear_btn = tk.Button(button_frame, text="Clear Fields", command=clear_fields, 
                      bg="orange", fg="white", font=("Arial", 10), padx=20, pady=5)
clear_btn.grid(row=0, column=1, padx=10)

# Add status bar
status_label = tk.Label(root, text="Ready to verify", relief=tk.SUNKEN, anchor=tk.W, 
                        font=("Arial", 8), fg="gray")
status_label.pack(side=tk.BOTTOM, fill=tk.X)

# Bind Enter key to verify function
root.bind('<Return>', lambda event: check_employee())

# Run the application
root.mainloop()
```
